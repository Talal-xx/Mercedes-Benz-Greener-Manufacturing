### For any column(s), the variance is equal to zero, then you need to remove those variable(s).

```python
## I'm going to use a very handy labrary to handle or to remove the feature which has 0 variance
from sklearn.feature_selection import VarianceThreshold
var_thr = VarianceThreshold(threshold = 0)
var_thr.fit(train_df.iloc[:,10:])
```

```python
len(var_thr.get_support())
```

```python
constant_col = [col for col in train_df.iloc[:,10:].columns
               if col not in train_df.iloc[:,10:].columns[var_thr.get_support()]]
```

```python
constant_col
```

```python
train_df.drop(columns=constant_col, inplace=True)
```

### Apply label encoder.

```python
train_df.head()
```

**First we must find the columns that contain objects**

```python
object_col = train_df.select_dtypes(include=[object])
object_col = object_col.columns
```

```python
object_col  # Shows the catagorical features
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

#### Creating loop to go through the desired columns and apply label encoder on them

```python
for i in object_col:
    train_df[i] = encoder.fit_transform(train_df[i])
```

```python
train_df.head()
```

#### I'll apply the same processes with the test dataset.

```python
object_test_col = test_df.select_dtypes(include=[object])
object_test_col = object_test_col.columns
```

```python
object_test_col
```

```python
from sklearn.preprocessing import LabelEncoder
test_encoder = LabelEncoder()
```

```python
# Creating loop to go through the desired columns and apply label encoder on them
for i in object_test_col:
    test_df[i] = test_encoder.fit_transform(test_df[i])
```

### Perform dimensionality reduction

```python
# Dim reducation using PCA
from sklearn.decomposition import PCA
pca = PCA(n_components= 2)
X_trian_transformed = pca.fit_transform(X_train)
X_test_transformed = pca.fit_transform(X_test)
explained_varience = pca.explained_variance_ratio_
```
